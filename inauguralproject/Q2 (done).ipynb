```python
%load_ext autoreload
%autoreload 2

import HouseholdSpecializationModel as HSM
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt # baseline module

# Predefine options for all plots
plt.rcParams.update({"axes.grid":True,"grid.color":"black","grid.alpha":"0.25","grid.linestyle":"-"})
plt.rcParams.update({'font.size': 14})
```

```python
konrad = HSM.HouseholdSpecializationModelClass()


sol = konrad.solve_discrete()

solution = konrad.solve_wF_vec(discrete = True, do_print=False)

# we extract our values from solutions
HF_vec = solution.HF_vec
HM_vec = solution.HM_vec
wF_vec = konrad.par.wF_vec

# we perform our desired log transformation
log_HFM = np.log(HF_vec/HM_vec)
log_wFM = np.log(wF_vec)  # note that wM = 1, so wF_vec/wM = wF_vec

# check what other parameters are than wF
print(konrad)

# create figure
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(log_HFM,log_wFM)
ax.set_title('Elasticity between ')
ax.set_xlabel('$log(HF/HM)$')
ax.set_ylabel('$log(wF)$')
```
